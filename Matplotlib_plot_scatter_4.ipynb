```python
# plot , scatter 
# 데이터 셋이 큰 경우는 plot이 더 효율적이다 왜?
# scatter는 각 점에 대한 설명을 나타낼 수 있어서 데이터가 크면 용량을 많이 잡아 먹는다.

rng = np.random.RandomState()
for marker in ['d','o','>']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
            label = "marker='{0}'".format(marker))
    plt.legend(numpoints=1)
    plt.xlim(0, 1.8);
```

```python
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha = 0.3,
           cmap = 'viridis')
plt.colorbar();
```
