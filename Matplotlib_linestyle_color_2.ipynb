```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#플롯 수정하기 : 선 색상과 스타일
x = np.linspace(0,10,1000)

plt.plot(x, np.sin(x - 0), color = 'blue') #색상을 이름으로 직접 지정
plt.plot(x, np.sin(x - 1), color = 'm') #짧은 색상 코드(rgbcmyk)
plt.plot(x, np.sin(x - 2), color = '0.35') #0과 1 사이로 회색조 지정(0에 가까울 수록 진해짐)

plt.plot(x, np.sin(x - 3), color = '#FFDD44') #16진수 코드 (RRGGBB, 00~FF 사이)
plt.plot(x, np.sin(x - 4), color = (1.0,0.2,0.3)) # RGB 튜플, 0과 1 값
plt.plot(x, np.sin(x - 5), color = 'chartreuse'); #모든 HTML 색상 이름을 지원
```

```python
#색상을 따로 지정하지 않으면 자동으로 다른 색상을 설정해준다.

plt.plot(x, np.sin(x))
plt.plot(x, np.sin(x-1))
plt.plot(x, np.sin(x-2))
```

```python
# 선 스타일

plt.plot(x, x + 0, linestyle = 'solid') #실선
plt.plot(x, x + 1, linestyle = 'dashed') #점선
plt.plot(x, x + 2, linestyle = 'dashdot') # 실선 사이 점
plt.plot(x, x + 3, linestyle = 'dotted') # 점

# 축약형

plt.plot(x, x + 7, linestyle = '-') # solid
plt.plot(x, x + 8, linestyle = '--') #dahsed
plt.plot(x, x + 9, linestyle = '-.') #dashdot
plt.plot(x, x + 10, linestyle = ':'); #dotted
```
