```python
import numpy as np
```

```python
fig = plt.figure() #그림 틀
ax = plt.axes() #축

#들어갈 값
x = np.linspace(-10,10, 1000) #(-10부터, 10까지) 
ax.plot(x, np.sin(x));
```

```python
fig = plt.figure() #그림 틀
ax = plt.axes() #축

#들어갈 값
x = np.linspace(-10,10, 1000) #(-10부터, 10까지) 
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```
