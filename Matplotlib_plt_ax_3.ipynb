```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100000) #start, end, num(개수) 즉, 여기서는 점의 개수니까 숫자가 높을 수록 정교해짐
plt.plot(x, np.sin(x))

plt.xlim(0,10) # x축 범위 지정
plt.ylim(10,-10) # y축 범위 역지정

plt.plot(x, np.sin(x - 3))
plt.axis([-1, 11, 9, -11]);
```

```python
plt.figure()

#xlim, ylim으로 따로 범위 지정하는 방법

plt.subplot(2,1,1) # (행, 열, 패널)
plt.plot(x, np.sin(x))
plt.title('A Sine Curve') # 타이틀
plt.xlim(0,10) # x축 범위 지정
plt.ylim(10,-10) # y축 범위 역지정
plt.xlabel('x') # y축 타이틀
plt.ylabel('sin(x)') # x축 타이틀

plt.plot(x, np.cos(x),'-',label='cos(x)') #범례 설정
plt.plot(x, np.sin(x),':',label='sin(x)')
plt.legend()

#"aixs"안에 리스트형태로 한 번에 지정하는 방법
plt.subplot(2,1,2)
plt.title('Example')
plt.plot(x, np.sin(x - 3))
plt.xlabel('example')
plt.ylabel('example')
plt.axis([-1, 11, 9, -11]); #+1, -1

#plt.axis('tight') 좁게, 'equal' 균형있게

# plt.plot?
```
