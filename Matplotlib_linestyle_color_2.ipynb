```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#  :   
x = np.linspace(0,10,1000)

plt.plot(x, np.sin(x - 0), color = 'blue') #   
plt.plot(x, np.sin(x - 1), color = 'm') #  (rgbcmyk)
plt.plot(x, np.sin(x - 2), color = '0.35') #0 1   (0   )

plt.plot(x, np.sin(x - 3), color = '#FFDD44') #16  (RRGGBB, 00~FF )
plt.plot(x, np.sin(x - 4), color = (1.0,0.2,0.3)) # RGB , 0 1 
plt.plot(x, np.sin(x - 5), color = 'chartreuse'); # HTML   
```

```python
#       .

plt.plot(x, np.sin(x))
plt.plot(x, np.sin(x-1))
plt.plot(x, np.sin(x-2))
```

```python
#  

plt.plot(x, x + 0, linestyle = 'solid') #
plt.plot(x, x + 1, linestyle = 'dashed') #
plt.plot(x, x + 2, linestyle = 'dashdot') #   
plt.plot(x, x + 3, linestyle = 'dotted') # 

# 

plt.plot(x, x + 7, linestyle = '-') # solid
plt.plot(x, x + 8, linestyle = '--') #dahsed
plt.plot(x, x + 9, linestyle = '-.') #dashdot
plt.plot(x, x + 10, linestyle = ':'); #dotted
```
